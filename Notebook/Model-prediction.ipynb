```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("./languages.csv", sep=";", header=None)
row_max_length = 70
row_min_length = 1
data
```

```python
def sumOfLength(row):
    i = 0
    for r in row:
        i += len(r)
    return i

def averageLength(row):
    words = row[1].split(",")
    #row["totalLength"] = sumOfLength(words)
    return sumOfLength(words) / row_max_length

def calcul_frequency_of_1(row):
    return row[1].count('1') / 70

def calcul_frequency_of_0(row):
    return row[1].count('0') / 70
```

```python
data["length"] = data[1].str.split(",").apply(lambda x: len(x))
```

```python
data["min_length"] = data[1].str.split(",").apply(lambda x: len(min(x, key=len)))
data["max_length"] = data[1].str.split(",").apply(lambda x: len(max(x, key=len)))
```

```python
data["average_length"] = data.apply(averageLength, axis = 1)
```

```python
data["1_frequency"] = data.apply(calcul_frequency_of_1, axis = 1)
data["0_frequency"] = data.apply(calcul_frequency_of_0, axis = 1)
```
