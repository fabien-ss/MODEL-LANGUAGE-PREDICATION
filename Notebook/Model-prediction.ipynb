Import all numpy and pandas

```python
import numpy as np
import pandas as pd
```

 Importing the data

```python
data = pd.read_csv("./languages.csv", sep=";", header=None)
row_max_length = 70
row_min_length = 1
data
data.columns = data.columns.astype(str)
```

Some function

```python
def sumOfLength(row):
    i = 0
    for r in row:
        i += len(r)
    return i

def averageLength(row):
    words = row["1"].split(",")
    #row["totalLength"] = sumOfLength(words)
    return sumOfLength(words) / row_max_length

def calcul_frequency_of_1(row):
    return row["1"].count('1') / 70

def calcul_frequency_of_0(row):
    return row["1"].count('0') / 70

data["length"] = data["1"].str.split(",").apply(lambda x: len(x))
```

Creation des colonnes

```python
data["min_length"] = data["1"].str.split(",").apply(lambda x: len(min(x, key=len)))
data["max_length"] = data["1"].str.split(",").apply(lambda x: len(max(x, key=len)))
```

```python
data["average_length"] = data.apply(averageLength, axis = 1)
```

```python
data["1_frequency"] = data.apply(calcul_frequency_of_1, axis = 1)
data["0_frequency"] = data.apply(calcul_frequency_of_0, axis = 1)
```

Division des donnÃ©es, code et non code

```python
codes = data.loc[(data["0"]) == 1]
not_codes = data.loc[(data["0"]) == 0]
not_codes
```

```python
f_codes = codes[:2500]
f_codes
```

Concatener les donnÃ©es code et non code de taille Ã©gales

```python
final_data = pd.concat([f_codes, f_not_codes])
```

```python
final_data.columns = final_data.columns.astype(str)

y_train = final_data.iloc[:, :1]
x_train = final_data.iloc[:, 2:]
```

Importer LogisticRegression

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

models = {}
```

    Cross validation

```python

model = LogisticRegression(max_iter=1000)  # Adjust max_iter if needed
scores = cross_val_score(model, x_train, y_train, cv=5, scoring='accuracy')
print(f"Mean Accuracy: {scores.mean():.2f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNC = KNeighborsClassifier(n_neighbors=10)

KNC.fit(x_train, y_train)
```

```python
def pre_processing(to_predict):
    prepare = pd.DataFrame([to_predict], columns=["1"])
    prepare["length"] = prepare["1"].str.split(",").apply(lambda x: len(x))
    prepare["min_length"] = prepare["1"].str.split(",").apply(lambda x: len(min(x, key=len)))
    prepare["max_length"] = prepare["1"].str.split(",").apply(lambda x: len(max(x, key=len)))
    prepare["average_length"] = prepare.apply(averageLength, axis = 1)
    prepare["1_frequency"] = prepare.apply(calcul_frequency_of_1, axis = 1)
    prepare["0_frequency"] = prepare.apply(calcul_frequency_of_0, axis = 1)
    # length
    # max_word_length
    # min_word_length
    # average_length
    # 1_frequency
    # 0_frequency
    return prepare
```

```python
import re

code_type = {
    1: {
        "status" : 200,
        "message": True,
    },
    0: {
        "status" : 200,
        "message": False
    }
}

def handle_api_request(code):
    print(code)
    regex = '^(0|1|,)*$'
    isValid = re.search(regex, code)
    if isValid == None:
        return {
            "status": 400,
            "message": "Not a valid code",
            "data": code
        }
    pre_proccessing = pre_processing(code)
    prediction = KNC.predict(pre_proccessing.iloc[:, 1:])
    code_t = code_type[prediction[0]]
    return code_t
```
